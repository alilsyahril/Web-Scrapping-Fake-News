## Scrapping News from Politifact Website
Website : https://www.politifact.com/factchecks/list/

### Background
Politifact is nonprofit website operated by the Poynter Institute in St. Petersburg, Florida, with offices there and in Washington, D.C. They are gathering every political statements from famous or influential individuals such as Donald Trump, Joe Biden, and Kamala Harris and verifying their statement is it true or false. This website can be useful to be a dataset for fake news classification

### 1. Trying for each elements

```python
# import libraries
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
import csv
```

```python
# Get the XTML file with requests module
url = 'https://www.politifact.com/factchecks/list/?page='
page = requests.get(url).text
print(page)
```

```python
# Coba lihat page kedua
page2 = requests.get(url+str(2)).text
print(page2)
```

```python
# trying to parse the first page using beautifulsoup html.parser
url = 'https://www.politifact.com/factchecks/list/?page='
page1 = requests.get(url+str(1)).text
soup1 = BeautifulSoup(page1, 'html.parser')
print(soup1)
```

```python
# find all with the prefix li (list) in the HTML file
items1 = soup1.findAll('li', 'o-listicle__item') # we can also use the sintax class_ = 'o-listicle__item'
print(items1)
```

### How to get each of the information we need

```python
# parsing only one statement using find with prefix li (list) in HTML file
item1 = soup1.find('li', 'o-listicle__item')
print(item1)
```

```python
# How to extract the statement
item1 = soup1.find('li', 'o-listicle__item')
statement1 = item1.find('div', class_ ='m-statement__quote').text.strip() # strip to remove all whitespaces
print(statement1)
```

```python
# How to get the label, we use the class c-image__original but we can use the class c-image__thumb as well
item1 = soup1.find('li', 'o-listicle__item')
label1 = item1.find('div', class_ ='m-statement__meter').find('img', class_ ='c-image__original').get('alt')
print(label1)
```

```python
# How to get the source/person
item1 = soup1.find('li', 'o-listicle__item')
source1 = item1.find('a', class_ ='m-statement__name').text.strip()
print(source1)
```

```python
# How to get the date, we are using regex
item1 = soup1.find('li', 'o-listicle__item')
date1 = re.findall('[a-zA-z]* \d{2}, \d{4} | [a-zA-z]* \d{1}, \d{4}', item1.find('div', class_ ='m-statement__desc').text.strip())[0]
print(date1) 
```

```python
# How to get the link
item1 = soup1.find('li', 'o-listicle__item')
link1 = 'https://www.politifact.com'
link1 += item1.find("div", class_ ='m-statement__quote').find('a')['href'].strip()
print(link1)
```

```python
# How to get the link version 2
item1 = soup1.find('li', 'o-listicle__item')
link1 = 'https://www.politifact.com'
link1 += item1.find("div",attrs={'class':'m-statement__quote'}).find('a')['href'].strip()
print(link1)
```

### 2. Do it in the first 10 pages

```python
datas = []
count_page = 0

for page in range(1, 11):
    count_page+=1
    print('scraping page :',count_page)
    req = requests.get(url+str(page))
    soup = BeautifulSoup(req.text, 'html.parser')
    news = soup.findAll('li', 'o-listicle__item')
    for new in news:
        statement = new.find('div', class_ ='m-statement__quote').text.strip()
        source = new.find('a', class_ ='m-statement__name').text.strip()
        try : date = re.findall('[a-zA-z]* \d{2}, \d{4} | [a-zA-z]* \d{1}, \d{4}', new.find('div', class_ ='m-statement__desc').text.strip())[0]
        except : date = ''
        label = new.find('div', class_ ='m-statement__meter').find('img', class_ ='c-image__original').get('alt')
        link = 'https://www.politifact.com'
        link += new.find("div",attrs={'class':'m-statement__quote'}).find('a')['href'].strip()
        # taruh di list
        datas.append([statement, source, date, label, link])
```

```python
print(datas)
```

```python
# Jadikan dataframe
headers = ['Statement', 'Source', 'Date', 'Label', 'Link']
df = pd.DataFrame(datas, columns=headers)
df
```

```python
# put it into a csv file
writer = csv.writer(open('fake_news_politifact.csv', 'w+', newline=''))
writer.writerow(headers)
for d in datas: writer.writerow(d)
```
